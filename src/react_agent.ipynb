```python
class InfraTools:

    @staticmethod
    def get_pod_status():
        return "CrashLoopBackOff"

    @staticmethod
    def check_disk():
        return "Disk usage is 97%"


class ReActAgent:

    def run(self):

        print("Thought: User reports pod crash. I should inspect pod status.\n")

        print("Action: get_pod_status()")
        pod_status = InfraTools.get_pod_status()
        print(f"Observation: {pod_status}\n")

        print("Thought: CrashLoopBackOff may be caused by storage issue. Check disk usage.\n")

        print("Action: check_disk()")
        disk_status = InfraTools.check_disk()
        print(f"Observation: {disk_status}\n")

        print("Thought: High disk utilization likely causing pod failure.\n")

        report = {
            "incident": "Kubernetes Pod Crash",
            "severity": "High",
            "root_cause": "Disk usage exceeded threshold",
            "pod_status": pod_status,
            "disk_status": disk_status,
            "recommendation": [
                "Clean unused logs",
                "Increase disk volume",
                "Restart affected pod"
            ]
        }

        print("Answer:")
        print(report)


agent = ReActAgent()
agent.run()
```
